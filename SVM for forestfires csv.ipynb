```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv('forestfires.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#### dropping month and day column
```

```python
df.drop(['month','day'],axis=1,inplace=True)
```

```python
df.head()
```

```python
predict=df.iloc[:,0:28]
target=df.iloc[:,28]
```

```python
predict
```

```python
target
```

```python
##### normalization
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return (x)
```

```python
fires = norm_func(predict)
```

```python
fires
```

```python
x_train,x_test,y_train,y_test=train_test_split(predict,target,test_size=0.3,stratify=target)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
##### kernel= linear
```

```python
model_linear=SVC(kernel='linear')
model_linear.fit(x_train,y_train)
```

```python
pred_test_linear=model_linear.predict(x_test)
```

```python
np.mean(pred_test_linear==y_test)
```

```python
acc = accuracy_score(y_test, pred_test_linear) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_test_linear)
```

```python
#### kernel=poly
```

```python
model_poly=SVC(kernel='poly')
model_poly.fit(x_train,y_train)
```

```python
pred_test_poly=model_poly.predict(x_test)
```

```python
np.mean(pred_test_poly==y_test)
```

```python
acc = accuracy_score(y_test, pred_test_poly) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_test_poly)
```

```python
#### kernel= sigmoid
```

```python
model_sigmoid=SVC(kernel='sigmoid')
model_sigmoid.fit(x_train,y_train)
```

```python
pred_test_sigmoid=model_sigmoid.predict(x_test)
```

```python
np.mean(pred_test_sigmoid==y_test)
```

```python
acc = accuracy_score(y_test, pred_test_sigmoid) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, pred_test_sigmoid)
```
